The sample contains 11,694,072 rows. I also verified that some (report_date, client_hash_id, content_hash_id) combinations appear more than once, so the sample includes duplicate records.

```python
con.sql("""
SELECT
    report_date,
    client_hash_id,
    content_hash_id,
    COUNT(*) AS rows
FROM read_parquet(
    'hf://datasets/FlyRank/internship-warehouse/fact_content_daily_performance_sample.parquet'
)
GROUP BY
    report_date,
    client_hash_id,
    content_hash_id
HAVING COUNT(*) > 1
LIMIT 10;
""").df()
```

## 3. Verify it with queries (grain, counts, missing values, windows)

*Every claim above gets a query cell here. A contract claim without a query next to it is a guess.*

```python
con.sql("""
SELECT
    MIN(report_date) AS start_date,
    MAX(report_date) AS end_date,
    COUNT(*) AS total_rows
FROM read_parquet(
    'hf://datasets/FlyRank/internship-warehouse/fact_content_daily_performance_sample.parquet'
);
""").df()
```

```python
con.sql("""
SELECT
    report_date,
    client_hash_id,
    content_hash_id,
    COUNT(*) AS rows
FROM read_parquet(
    'hf://datasets/FlyRank/internship-warehouse/fact_content_daily_performance_sample.parquet'
)
GROUP BY
    report_date,
    client_hash_id,
    content_hash_id
HAVING COUNT(*) > 1
LIMIT 10;
""").df()
```

```python
con.sql("""
SELECT
    SUM(CASE WHEN gsc_data_available = FALSE THEN 1 ELSE 0 END) AS gsc_missing,
    SUM(CASE WHEN ga4_data_available = FALSE THEN 1 ELSE 0 END) AS ga4_missing
FROM read_parquet(
    'hf://datasets/FlyRank/internship-warehouse/fact_content_daily_performance_sample.parquet'
);
""").df()
```

## 4. Data limits

*What can this data never tell you? Unbalanced history, GSC-only early rows, window overlaps.*

## Data limits

- This sample covers data only for the period from 2026-06-01 to 2026-06-30.
- The sample contains duplicate records for some combinations of report_date, client_hash_id, and content_hash_id.
- Some records have GSC or GA4 data marked as unavailable.
- Client and content identifiers are hashed, so they cannot be mapped back to the original identities.
- The sample dataset may not contain all historical or production data available in the complete warehouse.
